#### Observations:

  

The number of rating is 6 for quality of the wine. and maximum values are in 5,6,7. So we can say the wine is of average quality.

```python
# plot a pie and bar plot using subplots(1,2,figsize=(8,5))

fig,axes = plt.subplots(1,2,figsize=(15,5))
df['quality'].value_counts().plot(kind='bar',ax=axes[0])
df['quality'].value_counts().plot(kind='pie',ax=axes[1],autopct='%1.2f%%')
plt.show()
```

Observations:
  

The number of rating is 6 for quality of the wine. and maximum values are in 5,6,7. So we can say the wine is of average quality.
